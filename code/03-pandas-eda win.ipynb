### 시험 데이터 로드

```python
tst = pd.read_csv(tst_file, index_col=0)
print(tst.shape)
tst.head()
```

### 학습/시험 데이터 결합

```python
df = pd.concat([trn, tst], axis=0) # 위아래로 통합하겠다 axis를  0 , 옆으로는 1  
print(df.shape) # 320000 + 80000
df.tail()
```

```python
df.fillna(0, inplace=True) #NaN 과 같은 결측치를 0으로 채워줌 
df.tail()
```

### 피쳐 변환

```python
df['nObserve'].hist(bins=30)
```

```python
df['nObserve'] = df['nObserve'].apply(np.log1p)#해당 값에 1을 더하고 로그를 취해주는 함수(0인 값들을 넘어가기 위해)
# 그리고 다시 nObserve 라는 이름의 속성으로 바꾸어 넣겠다.
```

```python
df['nObserve'].hist(bins=30) #위의 과정은 첨도, 왜도가 커서 데이터가 한쪽에 쏠려있고 차이가 클 때 쓰는 방법으로 로그 그래프 형태에 따라 
# 좌측의 데이터를 많이 깍아서 주변 데이터의 분포를 더 잘보여주게 되며, 본래는 정규성을 가지도록하는 것이지만 아래 그래프를 보면 아직도 크게 치우쳐있음 
```

### 피쳐 생성

```python
# 관계가 있어보이는 것들은 차를 내서 살펴볼까?
df['d_dered_u'] = df['dered_u'] - df['u']
df['d_dered_g'] = df['dered_g'] - df['g']
df['d_dered_r'] = df['dered_r'] - df['r']
df['d_dered_i'] = df['dered_i'] - df['i']
df['d_dered_z'] = df['dered_z'] - df['z']
df['d_dered_rg'] = df['dered_r'] - df['dered_g']
df['d_dered_ig'] = df['dered_i'] - df['dered_g']
df['d_dered_zg'] = df['dered_z'] - df['dered_g']
df['d_dered_ri'] = df['dered_r'] - df['dered_i']
df['d_dered_rz'] = df['dered_r'] - df['dered_z']
df['d_dered_iz'] = df['dered_i'] - df['dered_z']
df['d_obs_det'] = df['nObserve'] - df['nDetect']
print(df.shape)
df.head()
```

### 피쳐 삭제

```python
df.corr().style.background_gradient() #상관 계숙 확인 함수 계수가 높으면 둘 중 하나만 쓰면 된다. 
```

```python
df.drop(['airmass_z', 'airmass_i', 'airmass_r', 'airmass_g', 'u', 'g', 'r', 'i', 'nDetect', 'd_dered_rg', 'd_dered_ri'], 
        axis=1, inplace=True)
print(df.shape)
df.head()
```

### 새로운 학습 데이터 파일 저장

```python
df.to_csv(feature_file)
```
